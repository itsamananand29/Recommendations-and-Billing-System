```python
import tkinter
from tkinter import *
import datetime
import time
from tkinter import messagebox
import sys 
import pandas as pd
import numpy as np
df1=pd.read_csv('GroceryStoreDataSet1.csv',header=None,na_filter=False)
df2=pd.read_csv('price.csv')
global df_data
col=['Name','Date','Item']
df_data=pd.DataFrame(columns=col)

d_current=datetime.datetime.now()
d=d_current.strftime('  %d-%B-20%y  ')
t=d_current.strftime('  %I:%M:%S  %p')
root=tkinter.Tk()
root.title('Product billing and recommendation')
global a,b,c,h,s,subamt
global prc,qt,itm,nm

qt=IntVar()
a,b,c,h=3,3,3,3
subamt=0
s=0
def tick():
    # get the current local time from the PC
    time_string = time.strftime('%H:%M:%S')
    # if time string has changed, update it
    clock.config(text=time_string)
    clock.after(200, tick)

    
def pp():
    global c,itm,prc,df_data
    if c==3:
        c+=1
        e4.delete(0,END)
        price=p.get()
        
        df_data=df_data.append({'Name':name.get(),'Date':d,'Item':price},ignore_index=True)
        for i in range(0,len(df2['Item'])):
            if df2['Item'].loc[i]==price:
                e4.insert(0,df2['Price'].loc[i])
                pr=df2['Price'].loc[i]
    else:
        ep=Entry(frame7,width=20,font=12,justify='center')
        ep.grid(row=a,column=0,pady=3)
        ep.delete(0,END)
        price=itm.get()
        df_data=df_data.append({'Name':name.get(),'Date':d,'Item':price},ignore_index=True)
        for i in range(0,len(df2['Item'])):
            if df2['Item'].loc[i]==price:
                ep.insert(0,df2['Price'].loc[i])
                prc=df2['Price'].loc[i]
    
def add_item():
    global a,b,c,h,prc,qt,subamt,itm
    pric=0
    qunt=IntVar()
        
    if b==3:
        
        pri=float(pr)
        qu=float(q.get())
        amt=pri*qu
        lamt=Label(frame7,text=amt,width=20,font=10,bg='white')
        lamt.grid(row=3,column=2,pady=2)
        b+=1 
        subamt=subamt+amt
           
    else:
        pc=float(prc)
        qnt=float(qt.get())
        amnt=pc*qnt
        la=Label(frame7,text=amnt,width=20,font=10,bg='white')
        la.grid(row=a,column=2,pady=3)
        subamt=subamt+amnt
    a+=1
    h+=1
    item=StringVar()
    item.set(df2['Item'].loc[0])
    prdct=OptionMenu(frame5,item,*it,command=lambda x:pp())
    prdct.config(width=16)
    prdct.grid(row=h,column=0)
    
    itm=item
    ep=Entry(frame7,width=20,font=12,justify='center')
    ep.grid(row=a,column=0,pady=3)
    
    price_pr=itm.get()
    for i in range(0,len(df2['Item'])):
        if df2['Item'].loc[i]==price_pr:
            ep.insert(0,df2['Price'].loc[i])
            pric=df2['Price'].loc[i]
    eq=Entry(frame7,textvariable=qunt,width=20,font=12,justify='center')
    eq.grid(row=a,column=1,pady=3)
    
    prc=pric
    qt=qunt
def subtotal():
    et.insert(0,subamt)
def find_price(*x):
    a_sum=0
    
    for j in x:
        
        for i in range(0,len(df2['Item'])):
            if df2['Item'].loc[i]==j:
                a_sum=a_sum+df2['Price'].loc[i]
                
    return a_sum    
    
def recom():
    
    import matplotlib as plt
    root1=tkinter.Tk()
    root1.geometry('270x450')
    %matplotlib inline
    
    def CurSelet(event):
        widget = event.widget
        selection=widget.curselection()
        picked = widget.get(selection[0])
        #print(len(picked))
        root2=tkinter.Tk()
        root2.title('Recommended')
        x=0
        for i in range(0,len(picked)):
            Label(root2,text=picked[i],font=40).grid(row=0,column=x)
            x=x+1
            Label(root2,text='+',font=40).grid(row=0,column=x)
            x=x+1
            
        a1=find_price(*picked)
        Label(root2,text='=',font=40).grid(row=0,column=x-1)
        Label(root2,text=a1,font=40).grid(row=0,column=x)
        root2.mainloop()
    
    
    
    from mlxtend.preprocessing import TransactionEncoder
    from mlxtend.frequent_patterns import apriori
    df1_ary=pd.np.array(df1)
    te=TransactionEncoder()
    te.fit(df1_ary)
    te_ary=te.transform(df1_ary)
    df=pd.DataFrame(te_ary,columns=te.columns_)
    df=df.drop('',axis=1)
    freq=apriori(df,min_support=0.05,use_colnames=True)
    freq['length']=freq['itemsets'].apply(lambda x:len(x))
    rs=freq[(freq['length']>=2) & (freq['support']>0.05)]
    l=len(rs)
    b=['']*l
    res=rs['itemsets']
    res.index=b
    r=pd.np.array(res)
    for i in range(0,l):
        r[i]=list(r[i])
    root1.title('Recommendations')
    listbox=Listbox(root1,width=40,height=80,selectmode=True,font=12)
    sb=Scrollbar(root1,orient='vertical',command=listbox.yview)
    sb.pack(side="right", fill="y")
    listbox.config(yscrollcommand=sb.set)
    sb.config(command=listbox.yview)
    listbox.bind('<<ListboxSelect>>',CurSelet)
    
    for i in range(0,len(res)):
        listbox.insert(i,r[i])
    listbox.pack()
    root1.mainloop()

def mbox():
   
    mess=messagebox.askokcancel(title='Submit',message='Are you sure you want to Submit!!')
    if mess==1:
        df_data.to_csv("R&B.csv",index=False,mode='a',header=False)
        messagebox.showinfo(title='Saved',message='Submited')


    
frame=Frame(root)
frame.grid(row=0,column=0)
frame1=Frame(root)
frame1.grid(row=1,column=0)
frame2=Frame(frame1)
frame2.grid(row=1,column=0,sticky='N')
frame3=Frame(frame1)
frame3.grid(padx=45,row=1,column=1,sticky='N')
frame4=Frame(frame2)
frame4.pack()
frame5=Frame(frame2)
frame5.pack(anchor=S)
frame6=Frame(frame3)
frame6.pack()
frame7=Frame(frame3)
frame7.pack(anchor=S)

l=Label(frame,text='Product Billing',fg='gray',font=('arial',26,'bold'))
l.pack(anchor=N)
l1=Label(frame4,text='Name',font=14,padx=10)
l1.grid(row=1,column=0)
name=StringVar()
e1=Entry(frame4,textvariable=name,width=20,font=14)
e1.grid(row=1,column=1)

prod=Label(frame5,text='Products',font=12)
prod.grid(row=2,column=0)
p=StringVar()

p.set(df2['Item'].loc[0])
it=[]
for i in df2['Item']:
    it.append(i)
prdt=OptionMenu(frame5,p,*it,command=lambda x:pp())
prdt.config(width=16)
prdt.grid(row=3,column=0)

l2=Label(frame6,text='Mobile Number',font=14)
l2.grid(row=1,column=0)
phno=StringVar()
e2=Entry(frame6,textvariable=phno,width=20,font=14)
e2.grid(row=1,column=1)
l3=Label(frame6,text='Date',font=14,padx=15,width=12)
l3.grid(row=1,column=3)
e7=Entry(frame6,width=13,font=14)
e7.grid(row=1,column=4)
e7.insert(0,d) 
l4=Label(frame6,text='Time',font=14,padx=15)
l4.grid(row=1,column=5)
#e8=Entry(frame3,width=12,font=14)
#e8.grid(row=1,column=6)
#e8.insert(0,t)
clock = Label(frame6, font=14, bg='white',fg='black')
clock.grid(row=1, column=6) 
tick()

price=Label(frame7,text='Price',font=12)
price.grid(row=2,column=0,sticky='N')
pr=IntVar()
e4=Entry(frame7,width=20,font=12,justify='center')
e4.grid(row=3,column=0,pady=3)

price_p=p.get()
for i in range(0,len(df2['Item'])):
    if df2['Item'].loc[i]==price_p:
        e4.insert(0,df2['Price'].loc[i])
        pr=df2['Price'].loc[i]


quantity=Label(frame7,text='Quantity',font=12)
quantity.grid(row=2,column=1,sticky='N')
q=IntVar()
e5=Entry(frame7,textvariable=q,width=20,font=12,justify='center')
e5.grid(row=3,column=1,pady=3)
amount=Label(frame7,text='Amount',font=12)
amount.grid(row=2,column=2,sticky='N')

lamt=Label(frame7,width=20,font=10,bg='white')
lamt.grid(row=3,column=2,pady=3)

et=Entry(frame7,width=14,justify='center',font=12)
et.grid(row=7,column=7)
b1=Button(frame7,text='Add +',width=15,padx=10,command = add_item).grid(row=3,column=7)
b2=Button(frame7,text='Submit',width=15,padx=10,command=mbox).grid(row=4,column=7)
b3=Button(frame7,text='Recommend',width=15,padx=10,command=recom).grid(row=5,column=7)
b4=Button(frame7,text='Total Bill',width=15,padx=10,command=subtotal).grid(row=6,column=7)

root.mainloop()

```
